## Polynomial of 6th degree

```python
import numpy as np
import sympy as sym
```

```python
e1 = 0.02544604382862076; e2 = 0.1292344072003028; e3 = 0.2970774243113014;
e4 = 0.5; e5 = 0.7029225756886985; e6 = 0.870765592799697; e7 = 0.974553956171379;
```

```python
A = np.array([[1, e1,  e1**2,  e1**3,  e1**4,  e1**5,  e1**6],[1, e2,  e2**2,  e2**3,  e2**4,  e2**5,  e2**6],
             [1, e3,  e3**2,  e3**3,  e3**4,  e3**5,  e3**6], [1, e4,  e4**2,  e4**3,  e4**4,  e4**5,  e4**6],
            [1, e5,  e5**2,  e5**3,  e5**4,  e5**5,  e5**6], [1, e6,  e6**2,  e6**3,  e6**4,  e6**5,  e6**6],
            [1, e7,  e7**2,  e7**3,  e7**4,  e7**5,  e7**6]])
```

```python
b1 = np.array([1, 0, 0, 0, 0, 0, 0]);
b2 = np.array([0, 1, 0, 0, 0, 0, 0]);
b3 = np.array([0, 0, 1, 0, 0, 0, 0]);
b4 = np.array([0, 0, 0, 1, 0, 0, 0]);
b5 = np.array([0, 0, 0, 0, 1, 0, 0]);
b6 = np.array([0, 0, 0, 0, 0, 1, 0]);
b7 = np.array([0, 0, 0, 0, 0, 0, 1]);
```

```python
# coefficients' polynomials
phi_1 = np.linalg.solve(A, b1);
phi_2 = np.linalg.solve(A, b2);
phi_3 = np.linalg.solve(A, b3);
phi_4 = np.linalg.solve(A, b4);
phi_5 = np.linalg.solve(A, b5);
phi_6 = np.linalg.solve(A, b6);
phi_7 = np.linalg.solve(A, b7);
phi_1
```
